```python
#Tasks 2: Data Cleaning
#( Ensure the dataset is clean for analysis.)
```

```python
#replacing space with "_"

df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
df.columns
```

```python
#check duplicate rows

df.duplicated().sum()
```

```python
#convert totalchange to numeric

df['totalcharges']=pd.to_numeric(df['totalcharges'], errors="coerce")
```

```python
#checking missing values

df["totalcharges"].isnull().sum()
```

```python
#remove missing values

df= df.dropna(subset=["totalcharges"]).reset_index(drop= True)
```

```python
#checking the tenure is negative or not as its also a data error

(df["tenure"]<0).sum()
```

```python
#Tasks 4: Customer Segmentation Visualization
```

```python
#create tenure group

df["tenure_group"]=pd.cut(
    df["tenure"],
    bins=[0,12,36,100],
    labels=["0-12 months", "13-36 months", "37+ months"])
```

```python
#count of customers in each group
df["tenure_group"].value_counts()
```

```python
#customer distribution by tenure group in donut chart 

tenure_counts = df['tenure_group'].value_counts()

plt.pie(
    tenure_counts.values,
    labels=tenure_counts.index
)

plt.title('Customer Distribution by Tenure Group')
plt.show()
```

```python
##customer distribution by tenure group in plotly express

tenure_counts = df['tenure_group'].value_counts().reset_index()

px.pie(
    tenure_counts,
    names='tenure_group',
    values='count',
    hole=0.4
)
```

```python
#customer distribution by tenure group in plotly object version

fig = go.Figure(
    go.Pie(
        labels=df['tenure_group'].value_counts().index,
        values=df['tenure_group'].value_counts().values,
        hole=0.5
    )
)
```
